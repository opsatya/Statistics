```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/sales_data_with_discounts.csv')
```

```python
data.head()
```

```python
#mean, mode, median, and standrand deviation

mean_cal = data['Total Sales Value'].mean()
print(mean_cal)

mode_cal = data['Avg Price'].mode()
print(mode_cal)

median_cal = data['Total Sales Value'].median()
print(median_cal)

standard_Cal = data['Net Sales Value'].std()
print(standard_Cal)

# print("\nSummary statistics for all numerical columns:")
print(data.describe())
```

```python
# Import matplotlib.pyplot if you haven't already
import matplotlib.pyplot as plt

# Select only the numerical columns for plotting histograms
# This is important because .hist() can only be applied to numerical data
#numerical_cols = data.select_dtypes(include=np.number).columns
numerical_cols = ['Total Sales Value', 'Avg Price', 'Net Sales Value', 'Discount Rate (%)', 'Discount Amount']

data[numerical_cols].hist(figsize=(15, 10))

# Display the plots
plt.tight_layout() # Adjust layout to prevent overlapping titles
plt.show()
```

```python
#Box plot

data.boxplot(column = 'Total Sales Value')

plt.tight_layout()
plt.show()
```

```python
# Calculate the average Avg Price for each Brand
avg_price_by_brand = data.groupby('Brand')['Avg Price'].mean()

    # Plot the average Avg Price per Brand
avg_price_by_brand.plot(kind='bar', figsize=(10, 6))
plt.title('Average Price per Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```
